```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Bissection function
def Bissection(f,a,b,Tol):
    try:
        assert(f(a)*f(b)<0)
        x0=(a+b)/2
        n=1
        while np.abs(f(x0))>=Tol:
            if np.abs(f(x0))==0:
                break
            elif np.sign(f(x0))==np.sign(f(b)):
                b=x0
                n+=1
            elif np.sign(f(x0))==np.sign(f(a)):
                a=x0
                n+=1
            x0=(a+b)/2    
        return x0,n
    except AssertionError:
        print('Bad input, same sign at both sides of the root.') 

# Secant function
def Secant(f,a,b,Tol):
    try:
        assert(f(a)*f(b)<0)
        n=1
        x_next=b-f(b)*(b-a)/(f(b)-f(a))
        while np.abs(f(x_next))>=Tol:
            a=b
            b=x_next
            x_next=b-f(b)*(b-a)/(f(b)-f(a))
            n+=1
        return x_next,n
    except AssertionError:
        print('Bad input, same sign at both sides of the root.')

# False position function
def F_position(f,a,b,Tol):
    try:
        assert(f(a)*f(b)<0)
        n=1
        x_next=b-f(b)*(b-a)/(f(b)-f(a))
        while np.abs(f(x_next))>=Tol:
            if f(a)*f(x_next)<0:
                b=x_next
            else:     
                a=x_next
            x_next=b-f(b)*(b-a)/(f(b)-f(a))
            n+=1
        return x_next,n
    except AssertionError:
        print('Bad input, same sign at both sides of the root.')

# Newton Raphson function
def NR_1d(f,x0,df,Tol):
    try:
        n=0
        while np.abs(f(x0))>=Tol:
            assert(df(x0)!=0)
            x0=x0-f(x0)/df(x0)
            n+=1
        return x0,n
    except AssertionError:
        print('df(x0)=0, try different initial guess.')
        
# Newton raphson function
def NR_multd(f,x0,J,Tol):
    try:
        n=1
        while np.linalg.norm(f(x0))>=Tol:
            assert(np.inv(J(x0))!=0)
            x0=x0-np.inv(J(x0))@f(x0)
            n+=1
        return x0,n
    except AssertionError:
        print('J(x0)=0, try different initial guess.')
        
# Parameters
Tol=10**-9
f=lambda x:2-x**3-np.sin(x)
df=lambda x:-3*x**2-np.cos(x)
a=0
b=5
x0=5

# Table of the data
solutions=[Bissection(f,a,b,Tol),Secant(f,a,b,Tol),
          F_position(f,a,b,Tol),NR_1d(f,x0,df,Tol)]
keys = ['Bissection', 'Secant', 'False position', 'Newton Raphson']
values = [value for value in solutions]
data = dict(zip(keys,values))
pd.DataFrame.from_dict(data, orient='index', columns=['x0', 'n'])                                                     
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Newton Raphson function
def NR(f,J,X0,Tol):
    while np.linalg.norm(f(X0))>Tol:
        X0=X0-np.linalg.inv(J(X0))@f(X0)
    return X0

# The functions f and g
f_g=lambda x: np.array([x[0]**6-15*x[0]**4*x[1]**2+15*x[0]**2*x[1]**4-
                        x[1]**6+x[0]**3-3*x[0]*x[1]**2-1,6*x[0]**5*x[1]-20*
                       x[0]**3*x[1]**3+6*x[0]*x[1]**5+3*x[0]**2*x[1]-
                        x[1]**3])

# Jaccobian
J=lambda x: np.array([[6*x[0]**5-60*x[0]**3*x[1]**2+30*x[0]*x[1]**4+3*x[0]
                       **2-3*x[1]**2,-30*x[0]**4*x[1]+60*x[0]**2*x[1]**3-
                       6*x[1]**5-6*x[0]*x[1]],[30*x[0]**4*x[1]-60*x[0]**2*
                        x[1]**3+6*x[1]**5+6*x[0]*x[1],6*x[0]**5-60*x[0]**
                        3*x[1]**2+30*x[0]*x[1]**4+3*x[0]**2-3*x[1]**2]])

limits=[2,0.2,0.02]
for k,limit in enumerate(limits):
    x=np.linspace(-limit,limit,num=400)
    y=x
    x0=np.zeros(2)
    Tol=10**-9
    root_mat=np.empty((np.size(x),np.size(y)))
    roots=np.array([[-1.74,0.85,-0.426,-0.426,0.587,0.587],
                    [0,0,0.737,-0.737,1.016,-1.016]])
    # Variables
    color_scale=np.arange(1,7,1)
    for i in range(0,np.size(x)):
        x0[0]=x[i]
        for j in range(0,np.size(y)):
            x0[1]=y[j]
            root=NR(f_g,J,x0,Tol)
            norms=np.array([np.linalg.norm(root-root_k) for
                            root_k in roots.T])
            ind=np.where(norms==np.min(norms,axis=0))
            root_mat[j,i]=int(color_scale[ind])
        
    # Plots
    plt.pcolormesh(x,y,root_mat)
    if k==0:
        plt.plot(roots[0,:],roots[1,:],'ro')
    plt.title('$z^6$+$z^3$-1')
    plt.xlabel('Re(z)')
    plt.ylabel('Im(z)')
    plt.colorbar()    
    plt.show()
```
