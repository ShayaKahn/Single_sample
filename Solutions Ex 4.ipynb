```python
import numpy as np
import matplotlib.pyplot as plt

h_vector=np.logspace(-15,-1,num=1000)
dfdx_Right=np.array([(np.sin(1+h)-np.sin(1))/h for h in h_vector])
dfdx_Symetric=np.array([(np.sin(1+h)-np.sin(1-h))/(2*h) for h in h_vector])
dfdx_analytical=np.cos(1)
relative_error_Right=np.abs((dfdx_Right-dfdx_analytical)/dfdx_analytical)
relative_error_Symetric=np.abs((dfdx_Symetric-dfdx_analytical)
                               /dfdx_analytical)

fig,axs=plt.subplots(2,1)
axs[0].loglog(h_vector,relative_error_Right)
axs[0].set_xlabel('h')
axs[0].set_ylabel('$E_r$')
axs[0].set_title('Relative error vs h in loglog graph-Right derivative')

axs[1].loglog(h_vector,relative_error_Symetric)
axs[1].set_xlabel('h')
axs[1].set_ylabel('$E_r$')
axs[1].set_title('Relative error vs h in loglog graph-Symetric derivative')

plt.subplots_adjust(bottom=0.1,top=1,wspace=0.4,hspace=0.4)
```
