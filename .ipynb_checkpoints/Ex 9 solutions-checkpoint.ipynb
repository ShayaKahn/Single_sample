```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.sparse import spdiags,csr_matrix

# Parameters
x_0=-5
L=20 
V_0=10 
k_0=2 
dx=0.1 
dt=0.005
alpha=1j*dt/(2*dx**2)
beta=1j*dt
tmax=2
t=np.arange(0,tmax,dt)
x=np.arange(-L,L,dx)
J=np.size(x)
T=np.size(t)
```

```python
# Simpsons rule function
def  Simpson(f,a,b,n):
     h=(b-a)/n
     I=0
     for i in range(0,n):
         I=I+1/6*h*(f(a+i*h)+4*f(a+i*h+h/2)+f(a+(i+1)*h))
     return I
```

```python
#Calculation of A:
f=lambda x: np.exp(-2*(x-x_0)**2)
I=Simpson(f,-L,L,10**4)
A=(1/I)**0.5

# Solution by Cranck nickolson
V=np.diag(V_0*np.exp(-0.5*x**2))
data=np.array([-2*np.ones(J),np.ones(J),np.ones(J)])
diags=np.array([0,1,-1])
S=spdiags(data,diags,J,J)
CN=csr_matrix(np.linalg.inv((np.eye(J)-alpha*S))@(np.eye(J)+alpha*S-beta*V))
psi_0 = A*np.exp(-(x-x_0)**2)*np.exp(1j*k_0*x)
psi=np.empty((J,T),dtype='complex')
psi[:,0]=psi_0
for n in range(1,np.size(t)):
    psi[:,n]=CN@psi[:,n-1]
# Plots
X,T=np.meshgrid(x,t)
fig,ax=plt.subplots()            
im=ax.pcolormesh(x,t,(np.abs(psi)**2).T)
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_title(r'$\|\psi(x,t)|^2$')
fig.colorbar(im, ax=ax)
```

```python
# Parameters
dx=0.1
dt=0.05 
x=np.arange(-10,10,dx)
t=np.arange(0,10,dt)
D=2
r=0.5
alpha=dt*D/(dx**2)
beta=dt*r
J=np.size(x)
T=np.size(t)
```

```python
# Solution by Operator spliting (CN + Euler).
data=np.array([-2*np.ones(J),np.ones(J),np.ones(J)])
diags=np.array([0,1,-1])
S=spdiags(data,diags,J,J).toarray()
S[0,1]=2
S[-1,-2]=2
CN=csr_matrix(np.linalg.inv((np.eye(J)-alpha*S))@(np.eye(J)+alpha*S))
psi=np.empty((J,T))
psi_0=np.exp(-x**2)
psi[:,0]=psi_0
for i in range(0,T-1):
    psi_temp=CN@psi[:,i]
    psi[:,i+1]=psi_temp+beta*psi_temp-beta*(psi_temp*psi_temp)
# Plots
fig,ax=plt.subplots()            
im=ax.pcolormesh(x,t,psi.T)
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_title(r'$\psi(x,t)$')
fig.colorbar(im, ax=ax)
```
