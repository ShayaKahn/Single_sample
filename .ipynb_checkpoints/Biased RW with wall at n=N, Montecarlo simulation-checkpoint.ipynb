```python
import numpy as np
import matplotlib.pyplot as plt

N=10
N_vector=np.arange(1, N, 1, dtype=int)
p=0.75
q=1-p
A=np.eye(N-1)-p*np.diag(np.ones(N-2),1)-q*np.diag(np.ones(N-2),-1)
b=np.concatenate((np.zeros(N-2),np.array([p])),axis=0)
p_n=np.linalg.solve(A, b)

plt.plot(N_vector, p_n,'bo-')
plt.xlabel('n',fontsize = 20)
plt.ylabel('$p_n$',fontsize = 20)
plt.title('p='+str(p)+' '+'q='+str(q))
```
