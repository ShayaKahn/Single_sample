```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

def trapezoidal(f,a,b,n):
    h=(b-a)/n
    J=0
    for i in range(1,n):
        J=J+f(a+i*h)
    I=h*((f(a)+f(b))/2+J)
    return I

def simpsons(f,a,b,n):
    h=(b-a)/n
    I=0
    for i in range(0,n):
        I=I+h/6*(f(a+i*h)+4*f(a+h/2+i*h)+f(a+(i+1)*h))
    return I   

# variables
a=0
b=1
n_vector=(2*np.ones((7,),dtype=int))**np.arange(2,9,1,dtype=int)
x=Symbol('x')
g=exp(x)
f=lambda x:np.exp(x)

def calc_analytic(f,x,a,b):
    I_analytical=integrate(f,(x,a,b))
    return float(I_analytical)

I_analytical = calc_analytic(g,x,a,b)
I_vector_trp = np.zeros(np.size(n_vector))
I_vector_smp = np.zeros(np.size(n_vector))
for i,n in enumerate(n_vector):
    I_vector_trp[i]=trapezoidal(f,a,b,n)
    I_vector_smp[i]=simpsons(f,a,b,n)     
    
# plots    
plt.loglog(n_vector,I_vector_trp-I_analytical,'bo-',label='trapezodial')
plt.loglog(n_vector,I_vector_smp-I_analytical,'ro-',label='simpsons')
plt.legend(loc='upper right')
plt.xlabel('n')
plt.ylabel('$\Delta$'+'n')
plt.show()
```
