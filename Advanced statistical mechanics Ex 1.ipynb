```python
from sympy import symbols
from sympy.plotting import plot
import matplotlib.pyplot as plt
q=symbols('q')
m=1
g=1
def set_init(qi,pi):
    p_plus=(pi**2-2*g*m**2*(q-qi))**0.5
    return p_plus
p1=plot(set_init(0,0),-set_init(0,0),
        set_init(1,1),-set_init(1,1),
        set_init(2,2),-set_init(2,2),
        title="Phase space", xlabel='q',
        ylabel='p',line_color= 'blue')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
N=1024
reals=1000000
s_array=np.empty(reals)
p_sat=1/1024*np.ones((1,N))
s_sat=np.sum(-p_sat*np.log(p_sat))
for i in range(0,reals):
    p=np.random.uniform(0,1,(1,N))
    p_set=p/np.sum(p)
    s=np.sum(-p_set*np.log(p_set))
    s_array[i]=s    
fig,ax=plt.subplots()    
ax.hist(s_array, bins='auto')
#ax.hist(np.array([s_sat]),color='red')
ax.set_title('Enthropy histogram')
plt.show()
```
